# 3D seismic modeling in homogeneous unbounded media

## Part I: Green's functions for isotropic media

### Displacement for vector force

The displacement field at the location $\mathbf{x}$ in the $n$-th direction caused by a point source $X_i(t)$ in the $i$-th direction placed at the origin is (Aki & Richards)

$$
u_n(\mathbf{x},t)=X_i*G_{ni}=\frac{1}{4\pi\rho}(3\gamma_n\gamma_i-\delta_{ni})\frac{1}{r^3}\int_{r/\alpha}^{r/\beta}\tau X_i(t-\tau)d\tau \\
+ \frac{1}{4\pi\rho\alpha^2}\gamma_n\gamma_i\frac{1}{r}X_i(t-\frac{r}{\alpha}) \\
- \frac{1}{4\pi\rho\beta^2}(\gamma_n\gamma_i-\delta_{ni})\frac{1}{r}X_i(t-\frac{r}{\beta}) \\
= \text{Near Field} + \text{Far Field P} + \text{Far Field S}
$$

where $r=|\mathbf{x}|$, $\gamma_i=x_i/r$ is the cosine angle with respect the $i$-th coordinate axis, $\delta_{ni}$ is the Kronecker delta, $\alpha$ and $\beta$ are the P- and S-wave velocities, and $\rho$ is the density.

### Displacement for a moment tensor source

The displacement field at the location $\mathbf{x}$ in the $n$-th direction caused by a "dipole" source $M_{ij}(t)$ with the force in the $i$-th direction and the "moment arm" in the $j$-th direction placed at the origin is (Aki & Richards)

$$
u_n(\mathbf{x},t)=M_{ij}*\frac{\partial{G_{ni}}}{\partial x_j}= \\
\frac{1}{4\pi\rho}(15\gamma_n\gamma_i\gamma_j-3\gamma_n\delta_{ij}-3\gamma_i\delta_{nj}-3\gamma_j\delta_{ni})\frac{1}{r^4}\int_{r/\alpha}^{r/\beta}\tau M_{ij}(t-\tau)d\tau \\
+ \frac{1}{4\pi\rho\alpha^2}\frac{6\gamma_n\gamma_i\gamma_j-\gamma_n\delta_{ij}-\gamma_i\delta_{nj}-\gamma_j\delta_{ni}}{r^2}M_{ij}(t-\frac{r}{\alpha})
- \frac{1}{4\pi\rho\beta^2}\frac{6\gamma_n\gamma_i\gamma_j-\gamma_n\delta_{ij}-\gamma_i\delta_{nj}-2\gamma_j\delta_{ni}}{r^2}M_{ij}(t-\frac{r}{\beta}) \\
+ \frac{1}{4\pi\rho\alpha^3}\frac{\gamma_n\gamma_i\gamma_j}{r}\dot{M}_{ij}(t-\frac{r}{\alpha}) \\
- \frac{1}{4\pi\rho\beta^3}\frac{\gamma_j(\gamma_n\gamma_i-\delta_{ni})}{r}\dot{M}_{ij}(t-\frac{r}{\beta}) \\
= \text{Near Field} + \text{Mid Field P} + \text{Mid Field S} + \text{Far Field P} + \text{Far Field S}
$$

### Strain for a moment tensor source

Refer to Luo et al. 2021 for normal strain, or to my dissertation, Appendix A for full expressions.

$$
u_{n,k} = \\ 
\frac{1}{4\pi\rho} (15(\gamma_{n,k}\gamma_i\gamma_j + \gamma_n\gamma_{i,k}\gamma_j + \gamma_n\gamma_i\gamma_{j,k}) - 3\gamma_{n,k}\delta_{ij} - 3\gamma_{i,k}\delta_{nj} - 3\gamma_{j,k}\delta_{ni})\frac{1}{r^4}\int_{r/\alpha}^{r/\beta}\tau M_{ij}(t-\tau)d\tau \\
+ \frac{1}{4\pi\rho}(15\gamma_n\gamma_i\gamma_j-3\gamma_n\delta_{ij}-3\gamma_i\delta_{nj}-3\gamma_j\delta_{ni}) (\frac{-4r_{,k}}{r^5}) \int_{r/\alpha}^{r/\beta}\tau M_{ij}(t-\tau)d\tau \\
+ \frac{1}{4\pi\rho}(15\gamma_n\gamma_i\gamma_j-3\gamma_n\delta_{ij}-3\gamma_i\delta_{nj}-3\gamma_j\delta_{ni}) \frac{1}{r^4} (\frac{r r_{,k}}{\beta^2}M_{ij}(t-\frac{r}{\beta}) - \frac{r r_{,k}}{\alpha^2}M_{ij}(t-\frac{r}{\alpha}) ) \\
+ \frac{1}{4\pi\rho\alpha^2}(\frac{r(6(\gamma_{n,k}\gamma_i\gamma_j + \gamma_n\gamma_{i,k}\gamma_j + \gamma_n\gamma_i\gamma_{j,k}) - \gamma_{n,k}\delta_{ij} - \gamma_{j,k}\delta_{nj} - \gamma_{j,k}\delta_{ni} ) - 2r_{,k}(6\gamma_n\gamma_i\gamma_j-\gamma_n\delta_{ij}-\gamma_i\delta_{nj}-\gamma_j\delta_{ni})}{r^3})M_{ij}(t-\frac{r}{\alpha}) \\ 
+ \frac{1}{4\pi\rho\alpha^2}\frac{6\gamma_n\gamma_i\gamma_j-\gamma_n\delta_{ij}-\gamma_i\delta_{nj}-\gamma_j\delta_{ni}}{r^2} (-\frac{r_{,k}}{\alpha})\dot{M}_{ij}(t-\frac{r}{\alpha}) \\
- \frac{1}{4\pi\rho\beta^2}(\frac{r(6(\gamma_{n,k}\gamma_i\gamma_j + \gamma_n\gamma_{i,k}\gamma_j + \gamma_n\gamma_i\gamma_{j,k}) - \gamma_{n,k}\delta_{ij} - \gamma_{j,k}\delta_{nj} - 2\gamma_{j,k}\delta_{ni} ) - 2r_{,k}(6\gamma_n\gamma_i\gamma_j-\gamma_n\delta_{ij}-\gamma_i\delta_{nj}-2\gamma_j\delta_{ni})}{r^3})M_{ij}(t-\frac{r}{\beta}) \\ 
- \frac{1}{4\pi\rho\beta^2}\frac{6\gamma_n\gamma_i\gamma_j-\gamma_n\delta_{ij}-\gamma_i\delta_{nj}-2\gamma_j\delta_{ni}}{r^2} (-\frac{r_{,k}}{\beta})\dot{M}_{ij}(t-\frac{r}{\beta}) \\ 
+ \frac{1}{4\pi\rho\alpha^3} (\frac{r(\gamma_{n,k}\gamma_i\gamma_j + \gamma_n\gamma_{i,k}\gamma_j + \gamma_n\gamma_i\gamma_{j,k}) - \gamma_n\gamma_i\gamma_j r_{,k}}{r^2}) \dot{M}_{ij}(t-\frac{r}{\alpha}) \\
+ \frac{1}{4\pi\rho\alpha^3}\frac{\gamma_n\gamma_i\gamma_j}{r} (-\frac{r_{,k}}{\alpha})\ddot{M}_{ij}(t-\frac{r}{\alpha}) \\
- \frac{1}{4\pi\rho\beta^3} (\frac{r(\gamma_{j,k}(\gamma_n\gamma_i-\delta_{ni}) + \gamma_j(\gamma_{n,k}\gamma_i + \gamma_n\gamma_{i,k})) - \gamma_j(\gamma_n\gamma_i-\delta_{ni}) r_{,k}}{r^2}) \dot{M}_{ij}(t-\frac{r}{\beta}) \\
- \frac{1}{4\pi\rho\beta^3}\frac{\gamma_j(\gamma_n\gamma_i-\delta_{ni})}{r} (-\frac{r_{,k}}{\beta})\ddot{M}_{ij}(t-\frac{r}{\beta})
$$

Then 

$$ 
\epsilon_{nk} = \frac{1}{2}(u_{n,k} + u_{k,n})
$$

where $u_{n,k}=\frac{\partial u_n}{\partial x_k}$, $\gamma_{i,k}=\frac{\partial \gamma_i}{\partial x_k}=\frac{1}{r}(\delta_{ik}-\frac{x_i x_k}{r^2})=\frac{1}{r}(\delta_{ik}-\gamma_i\gamma_k)$, and $r_{,k}=\frac{\partial r}{\partial x_k}=\frac{x_k}{r}=\gamma_k$.

## Part II: Green's functions for VTI media and vector force

There is no closed-form solution for general VTI media, but for simple anisotropy cases closed-form solutions were derived using higher-order ray theory (Vavrycuk, 2001). Below $\delta$, $\epsilon$, and $\gamma$ are the Thomsen anisotropy parameters.

### VTI type 1: $C_{13}=-C_{55}$ in the stiffness matrix

This case corresponds to $\delta=-\frac{1}{2}(1-\frac{\beta^2}{\alpha^2})$

$$
u_n(\mathbf{x},t)=X_i*G_{ni}=\frac{1}{4\pi\rho}\frac{1}{\sqrt{C_{55}/\rho}}\frac{g_{2n}g_{2i}-g_{3n}g_{3i}}{R^2}\int_{\tau_2}^{\tau_3} X_i(t-\tau)d\tau \\
+ \frac{1}{4\pi\rho}\frac{1}{C_{55}/\rho\sqrt{C_{33}/\rho}}\frac{g_{1n}g_{1i}}{\tau_1}X_i(t-\tau_1) \\
+ \frac{1}{4\pi\rho}\frac{1}{C_{11}/\rho\sqrt{C_{55}/\rho}}\frac{g_{2n}g_{2i}}{\tau_2}X_i(t-\tau_2) \\
+ \frac{1}{4\pi\rho}\frac{1}{C_{66}/\rho\sqrt{C_{55}/\rho}}\frac{g_{3n}g_{3i}}{\tau_3}X_i(t-\tau_3) \\
= \text{Near singularity} + \text{Far Field 1} + \text{Far Field 2} + \text{Far Field 3}
$$

where $\tau_1=\frac{r}{\sqrt{C_{55}/\rho}}\sqrt{\gamma_1^2+\gamma_2^2+\frac{C_{55}}{C_{33}}\gamma_3^2}$, $\tau_2=\frac{r}{\sqrt{C_{11}/\rho}}\sqrt{\gamma_1^2+\gamma_2^2+\frac{C_{11}}{C_{55}}\gamma_3^2}$, $\tau_3=\frac{r}{\sqrt{C_{66}/\rho}}\sqrt{\gamma_1^2+\gamma_2^2+\frac{C_{66}}{C_{55}}\gamma_3^2}$, $\mathbf{g}_1=(0,0,1)^T$, $\mathbf{g}_2=\frac{1}{R}(x_1,x_2,0)^T$, $\mathbf{g}_3=\frac{1}{R}(x_2,-x_1,0)^T$, $R^2=x_1^2 + x_2^2$.

### VTI type 2: $C_{33}=C_{11}$ and $C_{13}=C_{11}-2C_{55}$

This case corresponds to $\delta=\epsilon=0$

$$
u_n(\mathbf{x},t)=X_i*G_{ni}=\frac{1}{4\pi\rho}\frac{3g_{1n}g_{1i}-\delta_{ni}}{r^3}\int_{\tau_1}^{\tau_2} \tau X_i(t-\tau)d\tau \\
 + \frac{1}{4\pi\rho}\frac{1}{\sqrt{C_{55}/\rho}}\frac{g_{3n}^{\perp}g_{3i}^{\perp}-g_{3n}g_{3i}}{R^2}\int_{\tau_2}^{\tau_3} X_i(t-\tau)d\tau \\
+ \frac{1}{4\pi\rho}\frac{1}{C_{11}/\rho\sqrt{C_{11}/\rho}}\frac{g_{1n}g_{1i}}{\tau_1}X_i(t-\tau_1) \\
+ \frac{1}{4\pi\rho}\frac{1}{C_{55}/\rho\sqrt{C_{55}/\rho}}\frac{g_{2n}g_{2i}}{\tau_2}X_i(t-\tau_2) \\
+ \frac{1}{4\pi\rho}\frac{1}{C_{66}/\rho\sqrt{C_{55}/\rho}}\frac{g_{3n}g_{3i}}{\tau_3}X_i(t-\tau_3) \\
= \text{Near field} + \text{Near singularity} + \text{Far Field 1} + \text{Far Field 2} + \text{Far Field 3}
$$

where $\tau_1=\frac{r}{\sqrt{C_{11}/\rho}}$, $\tau_2=\frac{r}{\sqrt{C_{55}/\rho}}$, $\tau_3=\frac{r}{\sqrt{C_{66}/\rho}}\sqrt{\gamma_1^2+\gamma_2^2+\frac{C_{66}}{C_{55}}\gamma_3^2}$, $\mathbf{g}_1=\frac{1}{r}(x_1,x_2,x_3)^T$, $\mathbf{g}_2=-\frac{1}{rR}(-x_1x_3,-x_2x_3,R^2)^T$, $\mathbf{g}_3=\frac{1}{R}(x_2,-x_1,0)^T$, $\mathbf{g}_3^{\perp}=\frac{1}{R}(x_1,x_2,0)^T$.

### Analytical functions

```python
import scipy.integrate

def integrand1(tau, X0, t, dt):
    """Get the time function value at an arbitrary time t-tau (use linear interpolation)"""
    it = int((t-tau)/dt)
    nt = X0.shape[0]
    if it>=nt-1 or it<0:
        return 0
    else:
        return (tau-t+(it+1)*dt)/dt*X0[it] + (t-tau-it*dt)/dt*X0[it+1]
    
def integrand2(tau, X0, t, dt):
    """Get the time function value x tau at an arbitrary time t-tau (use linear interpolation)"""
    it = int((t-tau)/dt)
    nt = X0.shape[0]
    if it>=nt-1 or it<0:
        return 0
    else:
        return tau*((tau-t+(it+1)*dt)/dt*X0[it] + (t-tau-it*dt)/dt*X0[it+1])
    
def integrate1(X0,t,dt,a,b):
    """Integrate the array X0(t-tau) between a and b"""
    return scipy.integrate.quad(integrand1, a, b, args=(X0,t,dt),full_output=1)[0]

def integrate2(X0,t,dt,a,b):
    """Integrate the array tau*X0(t-tau) between a and b"""
    return scipy.integrate.quad(integrand2, a, b, args=(X0,t,dt),full_output=1)[0]
```

```python
# Analytical functions for displacement field with vector force in isotropic and VTI media
# 'i' is the force component, 'n' is the displacement component 
def near_field_iso_vf(X0,dt,alpha,beta,rho,x,i,n):
    """Near field displacement for a vector force source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gn=x[n]/r
    gi=x[i]/r
    dni=(n==i)
    a=r/alpha
    b=r/beta
    ia=round(a/dt)
    ib=round(b/dt)
    factor=1.0/(4*math.pi*rho)*(3*gn*gi-dni)/r**3
    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += itau*dt*X0[it-itau]*dt
        X[it] = val
        # X[it] = integrate2(X0,it*dt,dt,a,b)
    return factor*X

def far_field_iso_vf(X0,dt,alpha,beta,rho,x,i,n):
    """Far field displacement for a vector force source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gn=x[n]/r
    gi=x[i]/r
    dni=(n==i)
    a=r/alpha
    b=r/beta
    factor1=1.0/(4*math.pi*rho*alpha**2)*gn*gi/r
    factor2=1.0/(4*math.pi*rho*beta**2)*(gn*gi-dni)/r
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp1=nd.shift(temp1,a/dt,mode='nearest')
    temp2=nd.shift(temp2,b/dt,mode='nearest')
    X = factor1*temp1 - factor2*temp2
    return X

def total_field_iso_vf(X0,dt,alpha,beta,rho,x,i,n):
    """Total field displacement for a vector force source in isotropic media"""
    return near_field_iso_vf(X0,dt,alpha,beta,rho,x,n,i) + far_field_iso_vf(X0,dt,alpha,beta,rho,x,n,i)


# def near_singularity_vti1_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Near singularity displacement for a vector force source in VTI type 1 media"""
    mu=beta**2*rho
    lam=alpha**2*rho-2*mu
    c11=(1+2*epsilon)*(lam+2*mu)
    c33=lam+2*mu
    c55=mu
    c66=(1+2*gamma)*mu
    c13=math.sqrt(2*c33*(c33-c55)*delta+(c33-c55)**2)-c55
    c12=c11-2*c66
    
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))
    
    tau1=1.0/math.sqrt(c55/rho)*math.sqrt(x[0]**2+x[1]**2+c55/c33*x[2]**2)
    tau2=1.0/math.sqrt(c11/rho)*math.sqrt(x[0]**2+x[1]**2+c11/c55*x[2]**2)
    tau3=1.0/math.sqrt(c66/rho)*math.sqrt(x[0]**2+x[1]**2+c66/c55*x[2]**2)
    
    g1=np.array([0,0,1])
    g2=np.array([x[0],x[1],0])/R
    g3=np.array([x[1],-x[0],0])/R
    
    dni=(n==i)
    
    nt=X0.shape[0]
    X=np.zeros(nt)
        
    sign=1.0
    if tau2>tau3:
        tau2,tau3 = tau3,tau2
        sign=-1.0

    ia=round(tau2/dt)
    ib=round(tau3/dt)
    factor=sign*1.0/(4*math.pi*rho)*(g2[n]*g2[i]-g3[n]*g3[i])/(R**2*math.sqrt(c55/rho))
        
    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += X0[it-itau]*dt
        X[it] = factor*val
    return X

def near_singularity_vti1_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Near singularity displacement for a vector force source in VTI type 1 media"""
    vp0=alpha
    vs0=beta
    vp90=vp0*math.sqrt(1+2*epsilon)
    vsh90=vs0*math.sqrt(1+2*gamma)
       
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))

    gam=x/r
    d=np.eye(3)
    g=x/R

    Y=np.array([[0,0,1],[g[0],g[1],0],[g[1],-g[0],0]])

    u1 = gam[0]**2 + gam[1]**2 + (vs0/vp0*gam[2])**2
    u2 = gam[0]**2 + gam[1]**2 + (vp90/vs0*gam[2])**2
    u3 = gam[0]**2 + gam[1]**2 + (vsh90/vs0*gam[2])**2

    u = [u1, u2, u3]

    tau1=r/vs0*math.sqrt(u1)
    tau2=r/vp90*math.sqrt(u2)
    tau3=r/vsh90*math.sqrt(u3)

    tau=[tau1, tau2, tau3]

    RPns = Y[1,n]*Y[1,i]-Y[2,n]*Y[2,i]
    RP1ff = Y[0,n]*Y[0,i]/math.sqrt(u1)
    RP2ff = Y[1,n]*Y[1,i]/math.sqrt(u2)
    RP3ff = Y[2,n]*Y[2,i]/math.sqrt(u3)

    nt=X0.shape[0]
    X=np.zeros(nt)

    ia=round(tau[1]/dt)
    ib=round(tau[2]/dt)

    sign=1.0
    if tau[1]>tau[2]:
        ia,ib = ib,ia
        sign=-1.0

    factor = sign*RPns/(4*math.pi*rho*vs0*R**2)
    # factor = RPns/(4*math.pi*rho*vs0*R**2)

    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += X0[it-itau]*dt
        X[it] = val
        # X[it] = integrate1(X0,it*dt,dt,tau[1],tau[2])
    return factor*X


# def far_field_vti1_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Far field displacement for a vector force source in VTI type 1 media"""
    mu=beta**2*rho
    lam=alpha**2*rho-2*mu
    c11=(1+2*epsilon)*(lam+2*mu)
    c33=lam+2*mu
    c55=mu
    c66=(1+2*gamma)*mu
    c13=math.sqrt(2*c33*(c33-c55)*delta+(c33-c55)**2)-c55
    c12=c11-2*c66
    
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))
    
    tau1=1.0/math.sqrt(c55/rho)*math.sqrt(x[0]**2+x[1]**2+c55/c33*x[2]**2)
    tau2=1.0/math.sqrt(c11/rho)*math.sqrt(x[0]**2+x[1]**2+c11/c55*x[2]**2)
    tau3=1.0/math.sqrt(c66/rho)*math.sqrt(x[0]**2+x[1]**2+c66/c55*x[2]**2)
    
    g1=np.array([0,0,1])
    g2=np.array([x[0],x[1],0])/R
    g3=np.array([x[1],-x[0],0])/R
    
    dni=(n==i)
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    
    factor1=1.0/(4*math.pi*rho)*g1[n]*g1[i]/(tau1*c55/rho*math.sqrt(c33/rho))
    factor2=1.0/(4*math.pi*rho)*g2[n]*g2[i]/(tau2*c11/rho*math.sqrt(c55/rho))
    factor3=1.0/(4*math.pi*rho)*g3[n]*g3[i]/(tau3*c66/rho*math.sqrt(c55/rho))
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp3=np.copy(X0)
    temp1=nd.shift(temp1,tau1/dt,mode='nearest')
    temp2=nd.shift(temp2,tau2/dt,mode='nearest')
    temp3=nd.shift(temp3,tau3/dt,mode='nearest')
    X = factor1*temp1 + factor2*temp2 + factor3*temp3
    return X

def far_field_vti1_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Far field displacement for a vector force source in VTI type 1 media"""
    vp0=alpha
    vs0=beta
    vp90=vp0*math.sqrt(1+2*epsilon)
    vsh90=vs0*math.sqrt(1+2*gamma)
       
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))

    gam=x/r
    d=np.eye(3)
    g=x/R

    Y=np.array([[0,0,1],[g[0],g[1],0],[g[1],-g[0],0]])

    u1 = gam[0]**2 + gam[1]**2 + (vs0/vp0*gam[2])**2
    u2 = gam[0]**2 + gam[1]**2 + (vp90/vs0*gam[2])**2
    u3 = gam[0]**2 + gam[1]**2 + (vsh90/vs0*gam[2])**2

    u = [u1, u2, u3]

    tau1=r/vs0*math.sqrt(u1)
    tau2=r/vp90*math.sqrt(u2)
    tau3=r/vsh90*math.sqrt(u3)

    tau=[tau1, tau2, tau3]

    RPns = Y[1,n]*Y[1,i]-Y[2,n]*Y[2,i]
    RP1ff = Y[0,n]*Y[0,i]/math.sqrt(u1)
    RP2ff = Y[1,n]*Y[1,i]/math.sqrt(u2)
    RP3ff = Y[2,n]*Y[2,i]/math.sqrt(u3)

    factor1=RP1ff/(4*math.pi*rho*vs0*vp0*r)
    factor2=RP2ff/(4*math.pi*rho*vs0*vp90*r)
    factor3=RP3ff/(4*math.pi*rho*vs0*vsh90*r)
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp3=np.copy(X0)
    temp1=nd.shift(temp1,tau1/dt,mode='nearest')
    temp2=nd.shift(temp2,tau2/dt,mode='nearest')
    temp3=nd.shift(temp3,tau3/dt,mode='nearest')
    X = factor1*temp1 + factor2*temp2 + factor3*temp3
    return X

def total_field_vti1_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n):
    """Total field displacement for a vector force source in VTI type 1 media"""
    return near_singularity_vti1_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n) + far_field_vti1_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n)

def near_field_vti2_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Near field displacement for a vector force source in VTI type 2 media"""
    mu=beta**2*rho
    lam=alpha**2*rho-2*mu
    c11=(1+2*epsilon)*(lam+2*mu)
    c33=lam+2*mu
    c55=mu
    c66=(1+2*gamma)*mu
    c13=math.sqrt(2*c33*(c33-c55)*delta+(c33-c55)**2)-c55
    c12=c11-2*c66
    
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))
    
    tau1=r/math.sqrt(c11/rho)
    tau2=r/math.sqrt(c55/rho)
    tau3=1.0/math.sqrt(c66/rho)*math.sqrt(x[0]**2+x[1]**2+c66/c55*x[2]**2)
    
    g1=x/r
    g2=-np.array([-x[0]*x[2],-x[1]*x[2],R**2])/(r*R)
    g3=np.array([x[1],-x[0],0])/R
    g3perp=np.array([x[0],x[1],0])/R
    
    dni=(n==i)
    
    nt=X0.shape[0]
    X=np.zeros(nt)

    ia=round(tau1/dt)
    ib=round(tau2/dt)
    factor=1.0/(4*math.pi*rho)*(3*g1[n]*g1[i]-dni)/r**3
    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += itau*dt*X0[it-itau]*dt
        X[it] = factor*val
        # X[it] = integrate2(X0,it*dt,dt,tau1,tau2)
    return factor*X

def near_singularity_vti2_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Near singularity displacement for a vector force source in VTI type 2 media"""
    mu=beta**2*rho
    lam=alpha**2*rho-2*mu
    c11=(1+2*epsilon)*(lam+2*mu)
    c33=lam+2*mu
    c55=mu
    c66=(1+2*gamma)*mu
    c13=math.sqrt(2*c33*(c33-c55)*delta+(c33-c55)**2)-c55
    c12=c11-2*c66
    
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))
    
    tau1=r/math.sqrt(c11/rho)
    tau2=r/math.sqrt(c55/rho)
    tau3=1.0/math.sqrt(c66/rho)*math.sqrt(x[0]**2+x[1]**2+c66/c55*x[2]**2)
    
    g1=x/r
    g2=-np.array([-x[0]*x[2],-x[1]*x[2],R**2])/(r*R)
    g3=np.array([x[1],-x[0],0])/R
    g3perp=np.array([x[0],x[1],0])/R
    
    dni=(n==i)
    
    nt=X0.shape[0]
    X=np.zeros(nt)
        
    sign=1.0
    if tau2>tau3:
        tau2,tau3 = tau3,tau2
        sign=-1.0

    ia=round(tau2/dt)
    ib=round(tau3/dt)
    factor=sign*1.0/(4*math.pi*rho)*(g3perp[n]*g3perp[i]-g3[n]*g3[i])/(R**2*math.sqrt(c55/rho))
        
    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += X0[it-itau]*dt
        X[it] = factor*val
        # X[it] = integrate1(X0,it*dt,dt,tau2,tau3)
    return factor*X

def far_field_vti2_vf(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,n):
    """Far field displacement for a vector force source in VTI type 2 media"""
    mu=beta**2*rho
    lam=alpha**2*rho-2*mu
    c11=(1+2*epsilon)*(lam+2*mu)
    c33=lam+2*mu
    c55=mu
    c66=(1+2*gamma)*mu
    c13=math.sqrt(2*c33*(c33-c55)*delta+(c33-c55)**2)-c55
    c12=c11-2*c66
    
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))
    
    tau1=r/math.sqrt(c11/rho)
    tau2=r/math.sqrt(c55/rho)
    tau3=1.0/math.sqrt(c66/rho)*math.sqrt(x[0]**2+x[1]**2+c66/c55*x[2]**2)
    
    g1=x/r
    g2=-np.array([-x[0]*x[2],-x[1]*x[2],R**2])/(r*R)
    g3=np.array([x[1],-x[0],0])/R
    g3perp=np.array([x[0],x[1],0])/R
    
    dni=(n==i)
        
    nt=X0.shape[0]
    X=np.zeros(nt)
    
    ia=round(tau1/dt)
    ib=round(tau2/dt)
    ic=round(tau3/dt)
    factor1=1.0/(4*math.pi*rho)*g1[n]*g1[i]/(tau1*c11/rho*math.sqrt(c11/rho))
    factor2=1.0/(4*math.pi*rho)*g2[n]*g2[i]/(tau2*c55/rho*math.sqrt(c55/rho))
    factor3=1.0/(4*math.pi*rho)*g3[n]*g3[i]/(tau3*c66/rho*math.sqrt(c55/rho))
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp3=np.copy(X0)
    temp1=nd.shift(temp1,tau1/dt,mode='nearest')
    temp2=nd.shift(temp2,tau2/dt,mode='nearest')
    temp3=nd.shift(temp3,tau3/dt,mode='nearest')
    X = factor1*temp1 + factor2*temp2 + factor3*temp3
    return X

def total_field_vti2_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n):
    """Total field displacement for a vector force source in VTI type 2 media"""
    return near_field_vti2_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n) + near_singularity_vti2_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n) + far_field_vti2_vf(X0,dt,alpha,beta,rho,x,delta,epsilon,gamma,i,n)
```

```python
# Analytical functions for displacement and strain fields with moment tensor in isotropic media
# 'i,j' are the moment tensor components, 'n' is the displacement component, 'n,k' are the strain components
def near_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n):
    """Near field displacement for a moment tensor source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gi=x[i]/r
    gj=x[j]/r
    gn=x[n]/r
    dij=(i==j)
    dni=(n==i)
    dnj=(n==j)
    a=r/alpha
    b=r/beta
    ia=round(a/dt)
    ib=round(b/dt)
    factor=(15*gn*gi*gj-3*gn*dij-3*gi*dnj-3*gj*dni)/(4*math.pi*rho*r**4)
    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += itau*dt*X0[it-itau]*dt
        X[it] = val
        # X[it] = integrate2(X0,it*dt,dt,a,b)
    return factor*X

def mid_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n):
    """Mid field displacement for a moment tensor source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gi=x[i]/r
    gj=x[j]/r
    gn=x[n]/r
    dij=(i==j)
    dni=(n==i)
    dnj=(n==j)
    a=r/alpha
    b=r/beta
    factor1=(6*gn*gi*gj-gn*dij-gi*dnj-gj*dni)/(4*math.pi*rho*alpha**2*r**2)
    factor2=(6*gn*gi*gj-gn*dij-gi*dnj-2*gj*dni)/(4*math.pi*rho*beta**2*r**2)
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp1=nd.shift(temp1,a/dt,mode='nearest')
    temp2=nd.shift(temp2,b/dt,mode='nearest')
    X = factor1*temp1 - factor2*temp2
    return X

def far_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n):
    """Far field displacement for a moment tensor source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gi=x[i]/r
    gj=x[j]/r
    gn=x[n]/r
    dij=(i==j)
    dni=(n==i)
    dnj=(n==j)
    a=r/alpha
    b=r/beta
    factor1=gn*gi*gj/(4*math.pi*rho*alpha**3*r)
    factor2=gj*(gn*gi-dni)/(4*math.pi*rho*beta**3*r)
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp1=nd.shift(temp1,a/dt,mode='nearest')
    temp2=nd.shift(temp2,b/dt,mode='nearest')
    for it in range(1,nt-1):
        X[it] = factor1*0.5*(temp1[it+1]-temp1[it-1]) - factor2*0.5*(temp2[it+1]-temp2[it-1])
    X /= dt
    return X

def total_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n):
    """Total field displacement for a moment tensor source in isotropic media"""
    return near_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n) + mid_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n) + far_field_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n)

def geo_iso_mt(X0,M,dt,alpha,beta,rho,x):
    """Total field displacement vector for a full moment tensor source in isotropic media"""
    nt=X0.shape[0]
    geo=np.zeros((nt,3))
    for n in range(3):
        for p in range(3):
            for q in range(3):
                geo[:,n]+=total_field_iso_mt(M[p,q]*X0,dt,alpha,beta,rho,x,p,q,n)
    return geo

def near_singularity_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n):
    """Near singularity displacement for a moment tensor source in VTI type 1 media"""
    vp0=alpha
    vs0=beta
    vp90=vp0*math.sqrt(1+2*epsilon)
    vsh90=vs0*math.sqrt(1+2*gamma)
       
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))

    gam=x/r
    d=np.eye(3)
    g=x/R

    Y=np.array([[0,0,1],[g[0],g[1],0],[g[1],-g[0],0]])
    Y1=np.zeros((3,3))
    Y2=np.zeros((3,3))
    Y3=np.zeros((3,3))
    
    Y1[1,0]=d[0,0]-g[0]*g[0]*(1-d[2,0])
    Y1[1,1]=d[1,0]-g[1]*g[0]*(1-d[2,0])
    Y1[2,0]=Y1[1,1]
    Y1[2,1]=-Y1[1,0]

    Y2[1,0]=d[0,1]-g[0]*g[1]*(1-d[2,1])
    Y2[1,1]=d[1,1]-g[1]*g[1]*(1-d[2,1])
    Y2[2,0]=Y2[1,1]
    Y2[2,1]=-Y2[1,0]

    Y3[1,0]=d[0,2]-g[0]*g[2]*(1-d[2,2])
    Y3[1,1]=d[1,2]-g[1]*g[2]*(1-d[2,2])
    Y3[2,0]=Y3[1,1]
    Y3[2,1]=-Y3[1,0]

    Yj = [Y1, Y2, Y3]

    u1 = gam[0]**2 + gam[1]**2 + (vs0/vp0*gam[2])**2
    u2 = gam[0]**2 + gam[1]**2 + (vp90/vs0*gam[2])**2
    u3 = gam[0]**2 + gam[1]**2 + (vsh90/vs0*gam[2])**2

    u = [u1, u2, u3]

    u1_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vs0/vp0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u2_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vp90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u3_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vsh90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )

    u_s = [u1_s, u2_s, u3_s]

    tau1=r/vs0*math.sqrt(u1)
    tau2=r/vp90*math.sqrt(u2)
    tau3=r/vsh90*math.sqrt(u3)

    tau=[tau1, tau2, tau3]

    RPns = Y[1,n]*Y[1,i]-Y[2,n]*Y[2,i]
    RP1ff = Y[0,n]*Y[0,i]/math.sqrt(u1)
    RP2ff = Y[1,n]*Y[1,i]/math.sqrt(u2)
    RP3ff = Y[2,n]*Y[2,i]/math.sqrt(u3)

    dtau = np.zeros((3,3))
    for k in range(3):            
        dtau[0,k] = gam[k]/r * tau[0] + r/(vs0**2*tau[0]) * u_s[0]
        dtau[1,k] = gam[k]/r * tau[1] + r/(vp90**2*tau[1]) * u_s[1]
        dtau[2,k] = gam[k]/r * tau[2] + r/(vsh90**2*tau[2]) * u_s[2]

    RPns_s = Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i] - Yj[j][1,n]*Y[1,i]-Y[1,n]*Yj[j][1,i] + 2*g[j]*(1-d[2,j])*RPns
    RP1ff_s = Y[0,n]*Y[0,i]*u_s[0]/u[0]**(1.5) + gam[j]*RP1ff
    RP2ff_s = Y[1,n]*Y[1,i]*u_s[1]/u[1]**(1.5) - r/R*(Yj[j][1,n]*Y[1,i]+Y[1,n]*Yj[j][1,i])/math.sqrt(u[1]) + gam[j]*RP2ff
    RP3ff_s = Y[2,n]*Y[2,i]*u_s[2]/u[2]**(1.5) - r/R*(Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i])/math.sqrt(u[2]) + gam[j]*RP3ff

    nt=X0.shape[0]
    X=np.zeros(nt)

    ia=round(tau[1]/dt)
    ib=round(tau[2]/dt)

    sign=1.0
    if tau[1]>tau[2]:
        ia,ib = ib,ia
        sign=-1.0

    factor = sign*RPns_s/(4*math.pi*rho*vs0*R**3)

    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += X0[it-itau]*dt
        X[it] = val
        # X[it] = integrate1(X0,it*dt,dt,tau[1],tau[2])
    return factor*X

def mid_singularity_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n):
    """Mid singularity displacement for a moment tensor source in VTI type 1 media"""
    vp0=alpha
    vs0=beta
    vp90=vp0*math.sqrt(1+2*epsilon)
    vsh90=vs0*math.sqrt(1+2*gamma)
       
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))

    gam=x/r
    d=np.eye(3)
    g=x/R

    Y=np.array([[0,0,1],[g[0],g[1],0],[g[1],-g[0],0]])
    Y1=np.zeros((3,3))
    Y2=np.zeros((3,3))
    Y3=np.zeros((3,3))
    
    Y1[1,0]=d[0,0]-g[0]*g[0]*(1-d[2,0])
    Y1[1,1]=d[1,0]-g[1]*g[0]*(1-d[2,0])
    Y1[2,0]=Y1[1,1]
    Y1[2,1]=-Y1[1,0]

    Y2[1,0]=d[0,1]-g[0]*g[1]*(1-d[2,1])
    Y2[1,1]=d[1,1]-g[1]*g[1]*(1-d[2,1])
    Y2[2,0]=Y2[1,1]
    Y2[2,1]=-Y2[1,0]

    Y3[1,0]=d[0,2]-g[0]*g[2]*(1-d[2,2])
    Y3[1,1]=d[1,2]-g[1]*g[2]*(1-d[2,2])
    Y3[2,0]=Y3[1,1]
    Y3[2,1]=-Y3[1,0]

    Yj = [Y1, Y2, Y3]

    u1 = gam[0]**2 + gam[1]**2 + (vs0/vp0*gam[2])**2
    u2 = gam[0]**2 + gam[1]**2 + (vp90/vs0*gam[2])**2
    u3 = gam[0]**2 + gam[1]**2 + (vsh90/vs0*gam[2])**2

    u = [u1, u2, u3]

    u1_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vs0/vp0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u2_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vp90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u3_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vsh90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )

    u_s = [u1_s, u2_s, u3_s]

    tau1=r/vs0*math.sqrt(u1)
    tau2=r/vp90*math.sqrt(u2)
    tau3=r/vsh90*math.sqrt(u3)

    tau=[tau1, tau2, tau3]

    RPns = Y[1,n]*Y[1,i]-Y[2,n]*Y[2,i]
    RP1ff = Y[0,n]*Y[0,i]/math.sqrt(u1)
    RP2ff = Y[1,n]*Y[1,i]/math.sqrt(u2)
    RP3ff = Y[2,n]*Y[2,i]/math.sqrt(u3)

    dtau = np.zeros((3,3))
    for k in range(3):            
        dtau[0,k] = gam[k]/r * tau[0] + r/(vs0**2*tau[0]) * u_s[0]
        dtau[1,k] = gam[k]/r * tau[1] + r/(vp90**2*tau[1]) * u_s[1]
        dtau[2,k] = gam[k]/r * tau[2] + r/(vsh90**2*tau[2]) * u_s[2]

    RPns_s = Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i] - Yj[j][1,n]*Y[1,i]-Y[1,n]*Yj[j][1,i] + 2*g[j]*(1-d[2,j])*RPns
    RP1ff_s = Y[0,n]*Y[0,i]*u_s[0]/u[0]**(1.5) + gam[j]*RP1ff
    RP2ff_s = Y[1,n]*Y[1,i]*u_s[1]/u[1]**(1.5) - r/R*(Yj[j][1,n]*Y[1,i]+Y[1,n]*Yj[j][1,i])/math.sqrt(u[1]) + gam[j]*RP2ff
    RP3ff_s = Y[2,n]*Y[2,i]*u_s[2]/u[2]**(1.5) - r/R*(Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i])/math.sqrt(u[2]) + gam[j]*RP3ff

    nt=X0.shape[0]
    X=np.zeros(nt)

    temp2=np.copy(X0)
    temp3=np.copy(X0)
    temp2=nd.shift(temp2,tau[1]/dt,mode='nearest')
    temp3=nd.shift(temp3,tau[2]/dt,mode='nearest')

    factor = RPns/(4*math.pi*rho*vs0*R**2)

    X = factor*(dtau[1,j]*temp2 - dtau[2,j]*temp3)

    return X

def mid_field_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n):
    """Mid field displacement for a moment tensor source in VTI type 1 media"""
    vp0=alpha
    vs0=beta
    vp90=vp0*math.sqrt(1+2*epsilon)
    vsh90=vs0*math.sqrt(1+2*gamma)
       
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))

    gam=x/r
    d=np.eye(3)
    g=x/R

    Y=np.array([[0,0,1],[g[0],g[1],0],[g[1],-g[0],0]])
    Y1=np.zeros((3,3))
    Y2=np.zeros((3,3))
    Y3=np.zeros((3,3))
    
    Y1[1,0]=d[0,0]-g[0]*g[0]*(1-d[2,0])
    Y1[1,1]=d[1,0]-g[1]*g[0]*(1-d[2,0])
    Y1[2,0]=Y1[1,1]
    Y1[2,1]=-Y1[1,0]

    Y2[1,0]=d[0,1]-g[0]*g[1]*(1-d[2,1])
    Y2[1,1]=d[1,1]-g[1]*g[1]*(1-d[2,1])
    Y2[2,0]=Y2[1,1]
    Y2[2,1]=-Y2[1,0]

    Y3[1,0]=d[0,2]-g[0]*g[2]*(1-d[2,2])
    Y3[1,1]=d[1,2]-g[1]*g[2]*(1-d[2,2])
    Y3[2,0]=Y3[1,1]
    Y3[2,1]=-Y3[1,0]

    Yj = [Y1, Y2, Y3]

    u1 = gam[0]**2 + gam[1]**2 + (vs0/vp0*gam[2])**2
    u2 = gam[0]**2 + gam[1]**2 + (vp90/vs0*gam[2])**2
    u3 = gam[0]**2 + gam[1]**2 + (vsh90/vs0*gam[2])**2

    u = [u1, u2, u3]

    u1_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vs0/vp0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u2_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vp90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u3_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vsh90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )

    u_s = [u1_s, u2_s, u3_s]

    tau1=r/vs0*math.sqrt(u1)
    tau2=r/vp90*math.sqrt(u2)
    tau3=r/vsh90*math.sqrt(u3)

    tau=[tau1, tau2, tau3]

    RPns = Y[1,n]*Y[1,i]-Y[2,n]*Y[2,i]
    RP1ff = Y[0,n]*Y[0,i]/math.sqrt(u1)
    RP2ff = Y[1,n]*Y[1,i]/math.sqrt(u2)
    RP3ff = Y[2,n]*Y[2,i]/math.sqrt(u3)

    dtau = np.zeros((3,3))
    for k in range(3):            
        dtau[0,k] = gam[k]/r * tau[0] + r/(vs0**2*tau[0]) * u_s[0]
        dtau[1,k] = gam[k]/r * tau[1] + r/(vp90**2*tau[1]) * u_s[1]
        dtau[2,k] = gam[k]/r * tau[2] + r/(vsh90**2*tau[2]) * u_s[2]

    RPns_s = Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i] - Yj[j][1,n]*Y[1,i]-Y[1,n]*Yj[j][1,i] + 2*g[j]*(1-d[2,j])*RPns
    RP1ff_s = Y[0,n]*Y[0,i]*u_s[0]/u[0]**(1.5) + gam[j]*RP1ff
    RP2ff_s = Y[1,n]*Y[1,i]*u_s[1]/u[1]**(1.5) - r/R*(Yj[j][1,n]*Y[1,i]+Y[1,n]*Yj[j][1,i])/math.sqrt(u[1]) + gam[j]*RP2ff
    RP3ff_s = Y[2,n]*Y[2,i]*u_s[2]/u[2]**(1.5) - r/R*(Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i])/math.sqrt(u[2]) + gam[j]*RP3ff

    nt=X0.shape[0]
    X=np.zeros(nt)

    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp3=np.copy(X0)
    temp1=nd.shift(temp1,tau[0]/dt,mode='nearest')
    temp2=nd.shift(temp2,tau[1]/dt,mode='nearest')
    temp3=nd.shift(temp3,tau[2]/dt,mode='nearest')

    factor1 = RP1ff_s/(4*math.pi*rho*vp0*vs0*r**2)
    factor2 = RP2ff_s/(4*math.pi*rho*vp90*vs0*r**2)
    factor3 = RP3ff_s/(4*math.pi*rho*vsh90*vs0*r**2)

    X = factor1*temp1 + factor2*temp2 + factor3*temp3
    return X

def far_field_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n):
    """Far field displacement for a moment tensor source in VTI type 1 media"""
    vp0=alpha
    vs0=beta
    vp90=vp0*math.sqrt(1+2*epsilon)
    vsh90=vs0*math.sqrt(1+2*gamma)
       
    r=float(np.linalg.norm(x))
    R=float(np.linalg.norm(x[0:2]))

    gam=x/r
    d=np.eye(3)
    g=x/R

    Y=np.array([[0,0,1],[g[0],g[1],0],[g[1],-g[0],0]])
    Y1=np.zeros((3,3))
    Y2=np.zeros((3,3))
    Y3=np.zeros((3,3))
    
    Y1[1,0]=d[0,0]-g[0]*g[0]*(1-d[2,0])
    Y1[1,1]=d[1,0]-g[1]*g[0]*(1-d[2,0])
    Y1[2,0]=Y1[1,1]
    Y1[2,1]=-Y1[1,0]

    Y2[1,0]=d[0,1]-g[0]*g[1]*(1-d[2,1])
    Y2[1,1]=d[1,1]-g[1]*g[1]*(1-d[2,1])
    Y2[2,0]=Y2[1,1]
    Y2[2,1]=-Y2[1,0]

    Y3[1,0]=d[0,2]-g[0]*g[2]*(1-d[2,2])
    Y3[1,1]=d[1,2]-g[1]*g[2]*(1-d[2,2])
    Y3[2,0]=Y3[1,1]
    Y3[2,1]=-Y3[1,0]

    Yj = [Y1, Y2, Y3]

    u1 = gam[0]**2 + gam[1]**2 + (vs0/vp0*gam[2])**2
    u2 = gam[0]**2 + gam[1]**2 + (vp90/vs0*gam[2])**2
    u3 = gam[0]**2 + gam[1]**2 + (vsh90/vs0*gam[2])**2

    u = [u1, u2, u3]

    u1_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vs0/vp0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u2_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vp90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )
    u3_s = (gam[0]*(d[0,j]-gam[0]*gam[j]) + gam[1]*(d[1,j]-gam[1]*gam[j]) + (vsh90/vs0)**2*gam[2]*(d[2,j]-gam[2]*gam[j]) )

    u_s = [u1_s, u2_s, u3_s]

    tau1=r/vs0*math.sqrt(u1)
    tau2=r/vp90*math.sqrt(u2)
    tau3=r/vsh90*math.sqrt(u3)

    tau=[tau1, tau2, tau3]

    RPns = Y[1,n]*Y[1,i]-Y[2,n]*Y[2,i]
    RP1ff = Y[0,n]*Y[0,i]/math.sqrt(u1)
    RP2ff = Y[1,n]*Y[1,i]/math.sqrt(u2)
    RP3ff = Y[2,n]*Y[2,i]/math.sqrt(u3)

    dtau = np.zeros((3,3))
    for k in range(3):            
        dtau[0,k] = gam[k]/r * tau[0] + r/(vs0**2*tau[0]) * u_s[0]
        dtau[1,k] = gam[k]/r * tau[1] + r/(vp90**2*tau[1]) * u_s[1]
        dtau[2,k] = gam[k]/r * tau[2] + r/(vsh90**2*tau[2]) * u_s[2]

    RPns_s = Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i] - Yj[j][1,n]*Y[1,i]-Y[1,n]*Yj[j][1,i] + 2*g[j]*(1-d[2,j])*RPns
    RP1ff_s = Y[0,n]*Y[0,i]*u_s[0]/u[0]**(1.5) + gam[j]*RP1ff
    RP2ff_s = Y[1,n]*Y[1,i]*u_s[1]/u[1]**(1.5) - r/R*(Yj[j][1,n]*Y[1,i]+Y[1,n]*Yj[j][1,i])/math.sqrt(u[1]) + gam[j]*RP2ff
    RP3ff_s = Y[2,n]*Y[2,i]*u_s[2]/u[2]**(1.5) - r/R*(Yj[j][2,n]*Y[2,i]+Y[2,n]*Yj[j][2,i])/math.sqrt(u[2]) + gam[j]*RP3ff

    nt=X0.shape[0]
    X=np.zeros(nt)

    factor1 = RP1ff*dtau[0,j]/(4*math.pi*rho*vp0*vs0*r)
    factor2 = RP2ff*dtau[1,j]/(4*math.pi*rho*vp90*vs0*r)
    factor3 = RP3ff*dtau[2,j]/(4*math.pi*rho*vsh90*vs0*r)

    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp3=np.copy(X0)
    temp1=nd.shift(temp1,tau[0]/dt,mode='nearest')
    temp2=nd.shift(temp2,tau[1]/dt,mode='nearest')
    temp3=nd.shift(temp3,tau[2]/dt,mode='nearest')
    for it in range(1,nt-1):
        X[it] = factor1*0.5*(temp1[it+1]-temp1[it-1]) + factor2*0.5*(temp2[it+1]-temp2[it-1]) + factor3*0.5*(temp3[it+1]-temp3[it-1])
    X /= dt
    return X

def total_field_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n):
    """Total field displacement for a moment tensor source in VTI type 1 media"""
    a = near_singularity_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n) \
    + mid_singularity_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n) \
    + mid_field_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n) \
    + far_field_vti1_mt(X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,i,j,n)
    return a

def geo_vti1_mt(X0,M,dt,alpha,beta,rho,delta,epsilon,gamma,x):
    """Total field displacement vector for a full moment tensor source in VTI type 1 media"""
    nt=X0.shape[0]
    geo=np.zeros((nt,3))
    for n in range(3):
        for p in range(3):
            for q in range(3):
                geo[:,n]+=total_field_vti1_mt(M[p,q]*X0,dt,alpha,beta,rho,delta,epsilon,gamma,x,p,q,n)
    return geo

def near_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k):
    """Near field strain s_nk for a moment tensor source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gi=x[i]/r
    gj=x[j]/r
    gn=x[n]/r
    gik=((i==k)*r**2 - x[i]*x[k])/r**3
    gjk=((j==k)*r**2 - x[j]*x[k])/r**3
    gnk=((n==k)*r**2 - x[n]*x[k])/r**3
    rk=x[k]/r
    dij=(i==j)
    dni=(n==i)
    dnj=(n==j)
    a=r/alpha
    b=r/beta
    ak=rk/alpha
    bk=rk/beta
    ia=round(a/dt)
    ib=round(b/dt)
    f1=15*gn*gi*gj-3*gn*dij-3*gi*dnj-3*gj*dni
    f1k=15*(gnk*gi*gj+gn*gik*gj+gn*gi*gjk) - 3*gnk*dij - 3*gik*dnj - 3*gjk*dni
    factor=f1/(4*math.pi*rho*r**4)
    factork=(f1k*r - 4*rk*f1)/(4*math.pi*rho*r**5)
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp1=nd.shift(temp1,a/dt,mode='nearest')
    temp2=nd.shift(temp2,b/dt,mode='nearest')
    for it in range(nt):
        val=0
        for itau in range(ia,ib):
            if (it-itau>=0) and (it-itau<nt):
                val += itau*dt*X0[it-itau]*dt
        X[it] = val
        # X[it] = integrate2(X0,it*dt,dt,a,b)
    X = factork*X + factor*x[k]*(temp2/beta**2 - temp1/alpha**2)
    return X

def mid_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k): 
    """Mid field strain s_nk for a moment tensor source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gi=x[i]/r
    gj=x[j]/r
    gn=x[n]/r
    gik=((i==k)*r**2 - x[i]*x[k])/r**3
    gjk=((j==k)*r**2 - x[j]*x[k])/r**3
    gnk=((n==k)*r**2 - x[n]*x[k])/r**3
    rk=x[k]/r
    dij=(i==j)
    dni=(n==i)
    dnj=(n==j)
    a=r/alpha
    b=r/beta
    ak=rk/alpha
    bk=rk/beta
    factor1=(6*gn*gi*gj-gn*dij-gi*dnj-gj*dni)/(4*math.pi*rho*(alpha*r)**2)
    factor2=(6*gn*gi*gj-gn*dij-gi*dnj-2*gj*dni)/(4*math.pi*rho*(beta*r)**2)
    factor1k=(r**2*(6*(gnk*gi*gj+gn*gik*gj+gn*gi*gjk) - gnk*dij - gik*dnj - gjk*dni) - 2*x[k]*(6*gn*gi*gj-gn*dij-gi*dnj-gj*dni))/(4*math.pi*rho*(alpha*r**2)**2)
    factor2k=(r**2*(6*(gnk*gi*gj+gn*gik*gj+gn*gi*gjk) - gnk*dij - gik*dnj - 2*gjk*dni) - 2*x[k]*(6*gn*gi*gj-gn*dij-gi*dnj-2*gj*dni))/(4*math.pi*rho*(beta*r**2)**2)
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp1=nd.shift(temp1,a/dt,mode='nearest')
    temp2=nd.shift(temp2,b/dt,mode='nearest')
    for it in range(1,nt-1):
        X[it] = factor1k*temp1[it] - factor1*ak*0.5*(temp1[it+1]-temp1[it-1])/dt
        X[it] = X[it] - factor2k*temp2[it] + factor2*bk*0.5*(temp2[it+1]-temp2[it-1])/dt
    return X

def far_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k):
    """Far field strain s_nk for a moment tensor source in isotropic media"""
    
    nt=X0.shape[0]
    X=np.zeros(nt)
    r=float(np.linalg.norm(x))
    gi=x[i]/r
    gj=x[j]/r
    gn=x[n]/r
    gik=((i==k)*r**2 - x[i]*x[k])/r**3
    gjk=((j==k)*r**2 - x[j]*x[k])/r**3
    gnk=((n==k)*r**2 - x[n]*x[k])/r**3
    rk=x[k]/r
    dij=(i==j)
    dni=(n==i)
    dnj=(n==j)
    a=r/alpha
    b=r/beta
    ak=rk/alpha
    bk=rk/beta
    factor1=gn*gi*gj/(4*math.pi*rho*r*alpha**3)
    factor2=gj*(gn*gi-dni)/(4*math.pi*rho*r*beta**3)
    factor1k=(r**2*(gnk*gi*gj + gn*gik*gj + gn*gi*gjk) - x[k]*gn*gi*gj)/(4*math.pi*rho*(r*alpha)**3)
    factor2k=(r**2*(gjk*(gn*gi-dni) + gj*(gnk*gi+gn*gik)) - x[k]*gj*(gn*gi-dni))/(4*math.pi*rho*(r*beta)**3)
    temp1=np.copy(X0)
    temp2=np.copy(X0)
    temp1=nd.shift(temp1,a/dt,mode='nearest')
    temp2=nd.shift(temp2,b/dt,mode='nearest')
    for it in range(1,nt-1):
        X[it] = factor1k*0.5*(temp1[it+1]-temp1[it-1])/dt - factor1*ak*(temp1[it+1]-2*temp1[it]+temp1[it-1])/dt**2
        X[it] = X[it] - factor2k*0.5*(temp2[it+1]-temp2[it-1])/dt + factor2*bk*(temp2[it+1]-2*temp2[it]+temp2[it-1])/dt**2
    return X

def total_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k):
    """Total field strain for a moment tensor source in isotropic media"""
    X = near_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k) + mid_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k) + far_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,n,k)
    X += near_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,k,n) + mid_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,k,n) + far_field_strain_iso_mt(X0,dt,alpha,beta,rho,x,i,j,k,n)
    return 0.5*X

def strain_iso_mt(X0,M,dt,alpha,beta,rho,x):
    """Total field strain tensor for a full moment tensor source in isotropic media"""
    nt=X0.shape[0]
    strain=np.zeros((nt,3,3))
    for n in range(3):
        for k in range(n,3):
            for p in range(3):
                for q in range(3):
                    strain[:,n,k] += total_field_strain_iso_mt(M[p,q]*X0,dt,alpha,beta,rho,x,p,q,n,k)
    # complete the symmetry
    strain[:,1,0]=strain[:,0,1]
    strain[:,2,0]=strain[:,0,2]
    strain[:,2,1]=strain[:,1,2]
    return strain

def strain_tensor(s11,s22,s33,s12,s13,s23):
    """Build the 3x3 strain tensor from the 6 independent components"""
    
    nt=s11.shape[0]
    e=np.zeros((nt,3,3))
    e[:,0,0] = s11
    e[:,1,1] = s22
    e[:,2,2] = s33
    e[:,0,1] = s12
    e[:,0,2] = s13
    e[:,1,2] = s23
    e[:,1,0] = e[:,0,1]
    e[:,2,0] = e[:,0,2]
    e[:,2,1] = e[:,1,2]
    
    return e

def directional_strain(e,l):
    """Compute the scalar strain in a given direction"""
    
    nt=e.shape[0]
    lel=np.zeros(nt)
    for it in range(nt):
        lel[it] = np.dot(l,np.matmul(e[it,:,:],l))
    
    return lel
```

### Analytical vs numerical for isotropic and VTI media

```python
# source parameters
nt=501 # number of time samples
dt=0.001 # sampling rate (sec)
f0=15. # peak frequency (Hz)
shift=100 # shift wavelet (number of samples)

# Vector force
f=np.zeros((3))
fangle=0.25
fazimuth=0.0
f[0]=math.cos(fangle)*math.cos(fazimuth)
f[1]=math.cos(fangle)*math.sin(fazimuth)
f[2]=math.sin(fangle)

# Moment tensor
M=np.zeros((3,3))
M[0,0]=1.
M[1,1]=1.
M[2,2]=1.
M[0,1]=1.
M[0,2]=1.
M[1,2]=1.

# complete the symmetry
M[1,0]=M[0,1]
M[2,0]=M[0,2]
M[2,1]=M[1,2]

# model parameters
alpha=3500. # P-wave velocity (m/sec)
beta=1500. # S-wave velocity (m/sec)
rho=2000. # density (kg/m3)
delta=-0.5*(1. - (beta/alpha)**2) # anisotropy
epsilon=0.25 # anisotropy
gamma=0.2 # anisotropy

# source wavelet
w=np.zeros(nt)
pi = math.pi
tshift=shift*dt
sig=math.sqrt(2)/(2*f0*pi)
t=np.linspace(0,(nt-1)*dt,nt)
w=(1.0 - ((t-tshift)/sig)**2)*np.exp( -0.5*((t-tshift)/sig)**2)

# elastic models
nx=201
ny=201
nz=201
model_iso=np.zeros((3,ny,nx,nz))
model_iso[0,:,:]=alpha/1000. # Vp in km/s
model_iso[1,:,:]=beta/1000. # Vs in km/s
model_iso[2,:,:]=rho/1000. # Density in g/cc

model_vti=np.zeros((6,ny,nx,nz))
model_vti[0,:,:]=alpha/1000. # Vp in km/s
model_vti[1,:,:]=beta/1000. # Vs in km/s
model_vti[2,:,:]=rho/1000. # Density in g/cc
model_vti[3,:,:]=delta # Thomsen delta
model_vti[4,:,:]=epsilon # Thomsen epsilon
model_vti[5,:,:]=gamma # Thomsen gamma
```

### Analytical verification of different DAS approximations

```python
def getFiberPoints(x,gl,bending=0.0,nquad=11):
    """Get the coordinates and directions of the points regularly distributed over the fiber gauge section
    given the coordinates of the center, the gauge length, and the bending in radians.
    Bending is the angle between the two vector tangent to the fiber at either ends of the gauge.
    The gauge is assumed circular in the x-z plane and horizontal for bending=0."""

    y=np.zeros((nquad,3))
    l=np.zeros((nquad,3))

    if nquad==1:
        y[0]=x
        l[0]=[1,0,0]
    else:
        theta=np.linspace(-bending/2., bending/2., nquad)
        l[:,0]=np.cos(theta)
        l[:,2]=-np.sin(theta)
        if (bending<1e-6):
            y[:,0]=np.linspace(x[0]-gl/2.,x[0]+gl/2.,nquad)
            y[:,1]=x[1]
            y[:,2]=x[2]
        else:
            r = gl/bending
            center = np.copy(x)
            center[2] -= r
            y[:,0] = center[0] + r*np.cos(math.pi/2. - theta)
            y[:,1] = center[1]
            y[:,2] = center[2] + r*np.cos(theta)

    return y,l


def DAS_iso_mt(X0,M,dt,alpha,beta,rho,y,l,gl,bending=0):
    """DAS measurement in isotropic medium with moment tensor source.
    nquad is the size of y or l.
    nquad = 1: one point with one fiber direction.
    nquad = 2: two points with two fiber directions.
    nquad > 2: integral of strain over gauge length with nquad points."""

    nquad=y.shape[0]
    das = np.zeros(X0.shape)

    if nquad==1:
        xstart=y[0]-0.5*gl*l[0]
        xend=y[0]+0.5*gl*l[0]
        u = geo_iso_mt(X0,M,dt,alpha,beta,rho,xend) - geo_iso_mt(X0,M,dt,alpha,beta,rho,xstart)
        das = u[:,0]*l[0][0] + u[:,1]*l[0][1] + u[:,2]*l[0][2]
        das /= gl
    elif nquad==2:
        ua = geo_iso_mt(X0,M,dt,alpha,beta,rho,y[0])
        ub = geo_iso_mt(X0,M,dt,alpha,beta,rho,y[1])
        das = ub[:,0]*l[1][0] + ub[:,1]*l[1][1] + ub[:,2]*l[1][2] - (ua[:,0]*l[0][0] + ua[:,1]*l[0][1] + ua[:,2]*l[0][2])
        das /= gl
    else:
        ds=gl/(nquad -1.)
        for i in range(nquad):
            theta= i*ds*bending/gl - bending/2.
            u=geo_iso_mt(X0,M,dt,alpha,beta,rho,y[i])
            strain=strain_iso_mt(X0,M,dt,alpha,beta,rho,y[i])
            if i==0 or i==nquad-1:
                das[:] += 0.5 * (directional_strain(strain,l[i]) - bending/gl*(u[:,0]*math.sin(theta) + u[:,2]*math.cos(theta) ) )
            else:
                das[:] += (directional_strain(strain,l[i]) - bending/gl*(u[:,0]*math.sin(theta) + u[:,2]*math.cos(theta) ) )
        das /= (nquad-1.)
        
        # for i in range(nquad):
        #     dx=1.*dt*alpha*l[i]
        #     ub=geo_iso_mt(X0,M,dt,alpha,beta,rho,y[i]+dx)
        #     ua=geo_iso_mt(X0,M,dt,alpha,beta,rho,y[i]-dx)
        #     ul=(ub-ua)/(2. * np.linalg.norm(dx))
        #     ul = ul[:,0]*l[i][0] + ul[:,1]*l[i][1] + ul[:,2]*l[i][2]
        #     if i==0 or i==nquad-1:
        #         das[:] += 0.5 * ul
        #     else:
        #         das[:] +=ul
        # das /= (nquad-1.)

        # ua = geo_iso_mt(X0,M,dt,alpha,beta,rho,y[0])
        # ual = ua[:,0]*l[0][0] + ua[:,1]*l[0][1] + ua[:,2]*l[0][2]
        # for i in range(1,nquad):
        #     ub = geo_iso_mt(X0,M,dt,alpha,beta,rho,y[i])
        #     ubl = ub[:,0]*l[i][0] + ub[:,1]*l[i][1] + ub[:,2]*l[i][2]
        #     das += (ubl-ual)
        #     ual=ubl
        # das /= gl

    return das
```

```python
# DAS channel parameters
dip=0. # dip with respect to the horizontal (radians)
az=0. # azimuth with respect to x-axis (radians)
gl_long=50. # gauge length (m)
bending_deg=np.array([0., 1., 5., 20.]) # bending in degrees of the fiber
# bending_deg=np.array([0.]) # bending in degrees of the fiber
bending=bending_deg*math.pi/180. # bending in radians of the fiber

# wavelet parameters
nt_long=5*(nt-1) + 1
dt_long=dt/5.
t_long=np.linspace(0,(nt_long-1)*dt_long,nt_long)
lam = gl_long*np.array([0.25, 1., 10., 30.]) # wavelength in meters
# lam = gl_long*np.array([0.2]) # wavelength in meters
period=lam/alpha # period in seconds of the wavelet
w_mono=np.zeros((nt_long,lam.shape[0])) # monochromatic wavelet
for i in range(lam.shape[0]):
    w_mono[:,i] = np.sin(2*pi*t_long/period[i])

# isotropic moment tensor
M_iso=np.diag([1.,1.,1.])

# arrival time
t_arrival = np.linalg.norm(x-gl_long/2.1*np.array([1.,0.,0.])) / alpha

# DAS response
das_bent=np.zeros((nt_long,lam.shape[0],bending.shape[0]))
das_straight=np.zeros((nt_long,lam.shape[0],bending.shape[0]))
for b in range(bending.shape[0]):
    y1,l1 = getFiberPoints(x,gl_long,bending=bending[b],nquad=2)
    y2,l2 = getFiberPoints(x,gl_long,bending=bending[b],nquad=1)
    for per in range(period.shape[0]):
        das_bent[:,per,b] = DAS_iso_mt(w_mono[:,per],M_iso,dt,alpha,beta,rho,y1,l1,gl_long)
        das_straight[:,per,b] = DAS_iso_mt(w_mono[:,per],M_iso,dt,alpha,beta,rho,y2,l2,gl_long)
```

```python
# Assessment of straight fiber approximation

rcParams['font.size'] = 8
rcParams['font.family'] = 'sans-serif'

fig, ax = plt.subplots(lam.shape[0],bending.shape[0],figsize=(6.33, 6.0),dpi=300)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.1, hspace=0.25)

for i in range(lam.shape[0]):
    for j in range(bending.shape[0]):
        it1=int(round(t_arrival/dt_long))
        it2=int(round((t_arrival + 3*period[i])/dt_long))
        ax[i,j].plot(t_long[it1:it2],das_bent[it1:it2,i,j],     c='k', linewidth=1.5, label='Exact')
        ax[i,j].plot(t_long[it1:it2],das_straight[it1:it2,i,j], c='r', linewidth=0.7, label='Straight')
        ax[i,j].set_xlim([t_arrival,min(0.95*t[-1] , t_arrival + 3*period[i] ) ])
        if i==lam.shape[0]-1:
            ax[i,j].set_xlabel('Time (s)')
        if j==0:
            ax[i,j].set_ylabel('Strain (m/m)')
        # else:
            # ax[i,j].set_yticklabels([])
        # ax[i,j].legend(loc='upper left',prop={'size': 4})
        ax[i,j].tick_params(axis='both', which='major', labelsize=6)

ax[0,0].set_title( r'$ben=0^o$, $wl/gl=0.25$', loc='center',size=6,y=1, pad=12)
ax[0,1].set_title( r'$ben=1^o$, $wl/gl=0.25$', loc='center',size=6,y=1, pad=12)
ax[0,2].set_title( r'$ben=5^o$, $wl/gl=0.25$', loc='center',size=6,y=1, pad=12)
ax[0,3].set_title( r'$ben=20^o$,$wl/gl=0.25$', loc='center',size=6,y=1, pad=12)

ax[1,0].set_title( r'$ben=0^o$, $wl/gl=1$', loc='center',size=6,y=1, pad=12)
ax[1,1].set_title( r'$ben=1^o$, $wl/gl=1$', loc='center',size=6,y=1, pad=12)
ax[1,2].set_title( r'$ben=5^o$, $wl/gl=1$', loc='center',size=6,y=1, pad=12)
ax[1,3].set_title( r'$ben=20^o$,$wl/gl=1$', loc='center',size=6,y=1, pad=12)

ax[2,0].set_title( r'$ben=0^o$, $wl/gl=10$', loc='center',size=6,y=1, pad=12)
ax[2,1].set_title( r'$ben=1^o$, $wl/gl=10$', loc='center',size=6,y=1, pad=12)
ax[2,2].set_title( r'$ben=5^o$, $wl/gl=10$', loc='center',size=6,y=1, pad=12)
ax[2,3].set_title( r'$ben=20^o$, $wl/gl=10$', loc='center',size=6,y=1, pad=12)

ax[3,0].set_title( r'$ben=0^o$, $wl/gl=30$', loc='center',size=6,y=1, pad=12)
ax[3,1].set_title( r'$ben=1^o$, $wl/gl=30$', loc='center',size=6,y=1, pad=12)
ax[3,2].set_title( r'$ben=5^o$, $wl/gl=30$', loc='center',size=6,y=1, pad=12)
ax[3,3].set_title( r'$ben=20^o$, $wl/gl=30$', loc='center',size=6,y=1, pad=12)

plt.tight_layout()

error=np.zeros((lam.shape[0],bending.shape[0]))
for i in range(lam.shape[0]):
    for j in range(bending.shape[0]):
        it1=int(round(t_arrival/dt_long))
        it2=int(round((0.95*min(t[-1] , t_arrival + period[i] ))/dt_long))
        error[i,j] = 100*np.linalg.norm(das_straight[it1:it2,i,j]-das_bent[it1:it2,i,j])/np.linalg.norm(das_bent[it1:it2,i,j])
        print("Relative error per period with bending=%f and lam/gl=%f is %.1f%%" %(bending[j],lam[i]/gl_long,error[i,j]))

plt.savefig(figpath+'das_straight_errors.png',bbox_inches='tight',format='png')
```
